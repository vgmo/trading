# Low, High, and Close SMAs 1-10 days comparing vs close price

```python
# Hold all indicators in list (list of strings)
indicator_list = []
# Function to create sma
def create_sma(duration, price, df_input, indicator_list_input):
    col_name = 'SMA_' + str(duration) + '_' + price.lower()
    indicator_list_input.append(col_name)
    df_input[col_name] = df_input[price.lower()].rolling(duration, min_periods=1).mean()
    
# Loop through SMA function to create multiple SMAs
for i in range(1, 11):
    create_sma(i, 'low', df, indicator_list)
    create_sma(i, 'high', df, indicator_list)
    create_sma(i, 'close', df, indicator_list)
# Create 50 day SMA 
# df['SMA_50_low'] = df['low'].rolling(50, min_periods=1).mean()
# df['SMA_50_high'] = df['high'].rolling(50, min_periods=1).mean()
```

```python
def sma_backtest(indicator, cash, days, compare_price, start_range, df_input):
    buy_cash = 0
    status = False
    trade_data = []
    number_of_shares = 0
    ind_arr = indicator.split('_')
    # start_range = int(ind_arr[1])
    end_range = start_range + days
    for i in range(start_range, end_range):
         # At the end sell all positions and update remaining cash
        if i == end_range - 1:
            if status == True:
                cash = number_of_shares * df_input.iloc[i+1]['open']
                status = False
                number_of_shares = 0
                trade_data.append({'type': 'sell', 'price': df_input.iloc[i+1]['open'], 'cash': cash, 'shares': number_of_shares, 'profit': cash - buy_cash, 'percent_change': round((cash - buy_cash) / buy_cash, 2) * 100,'indicator': indicator, 'date': df_input.iloc[i]['date']})
                break
            else:
                trade_data.append({'type': 'hold', 'price': 0, 'cash': buy_cash, 'shares': 0, 'profit': 0, 'percent_change': 0,'indicator': indicator, 'date': df_input.iloc[i]['date']})
                break
                
        # Check if close is above moving average
        if df_input.iloc[i][compare_price.lower()] > df_input.iloc[i][indicator]:
            # If the close is above either buy or hold the security
            if status == False:
                # Buy the stock with the cash
                number_of_shares = cash / df_input.iloc[i+1]['open']
                buy_cash = cash
                cash = 0
                status = True
                trade_data.append({'type': 'buy', 'price': df_input.iloc[i+1]['open'], 'cash': cash, 'shares': number_of_shares, 'profit': 0, 'percent_change': 0,'indicator': indicator, 'date': df_input.iloc[i]['date'] })
        if df_input.iloc[i][compare_price.lower()] < df_input.iloc[i][indicator]:
            if status == True:
                cash = number_of_shares * df_input.iloc[i+1]['open']
                status = False
                number_of_shares = 0
                trade_data.append({'type': 'sell', 'price': df_input.iloc[i+1]['open'], 'cash': cash, 'shares': number_of_shares, 'profit': cash - buy_cash, 'percent_change': round((cash - buy_cash) / buy_cash, 2) * 100,'indicator': indicator, 'date': df_input.iloc[i]['date']})

       
    return pd.DataFrame(trade_data)
```

```python
# df_trade_data = sma_backtest('SMA_50_high', 1000, 20)
# df_trade_data_1 = sma_backtest('SMA_50_low', 1000, 20)

df_backtest = pd.DataFrame(columns=['type', 'price', 'cash', 'shares', 'profit', 'indicator', 'date'])
for i in indicator_list:
    df_backtest = pd.concat([df_backtest, sma_backtest(i, 1000, 400, 'close', 15, df)]) 

```

```python
def vwap(df):
    v = df['volume'].values
    tp = (df['low'] + df['close'] + df['high']).div(3).values
    df['vwap'] = ((tp * v).cumsum() / v.cumsum())
```
